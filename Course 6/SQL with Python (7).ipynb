## Join Statements

##### HR DATABASE TABLES: CREATE, UPDATE and DROP VIEW TABLES

##### Problem 1: Select the names and job start dates of all employees who work for the department number 5.

```python
%%sql 
SELECT E.F_NAME,E.L_NAME, JH.START_DATE 
FROM EMPLOYEES as E 
INNER JOIN JOB_HISTORY as JH on E.EMP_ID=JH.EMPL_ID 
WHERE E.DEP_ID ='5';  
```

##### Problem 2: Select the names, job start dates, and job titles of all employees who work for the department number 5.

```python
%%sql 
SELECT E.F_NAME,E.L_NAME, JH.START_DATE, J.JOB_TITLE
FROM EMPLOYEES E 
INNER JOIN JOB_HISTORY JH ON E.EMP_ID=JH.EMPL_ID 
INNER JOIN JOBS J ON JH.JOBS_ID=J.JOB_IDENT 
WHERE E.DEP_ID ='5';  
```

##### Problem 3: Perform a Left Outer Join on the EMPLOYEES and DEPARTMENT tables and select employee id, last name, department id and department name for all employees.

```python
%%sql 
SELECT E.EMP_ID, E.L_NAME, D.DEPT_ID_DEP, D.DEP_NAME
FROM EMPLOYEES E 
LEFT JOIN DEPARTMENTS D ON E.DEP_ID = D.DEPT_ID_DEP;
```

##### Problem 4: Re-write the previous query but limit the result set to include only the rows for employees born before 1980.

```python
%%sql 
SELECT E.EMP_ID, E.L_NAME, D.DEPT_ID_DEP, D.DEP_NAME
FROM EMPLOYEES E 
LEFT JOIN DEPARTMENTS D ON E.DEP_ID = D.DEPT_ID_DEP
WHERE YEAR(E.B_DATE) < 1980;
```

##### Problem 5: Re-write the previous query but have the result set include all the employees but department names for only the employees who were born before 1980.

```python
%%sql 
SELECT E.EMP_ID, E.L_NAME, D.DEPT_ID_DEP, D.DEP_NAME
FROM EMPLOYEES E 
LEFT JOIN DEPARTMENTS D ON E.DEP_ID = D.DEPT_ID_DEP
AND YEAR(E.B_DATE) < 1980;
```

##### Problem 6: Perform a Full Join on the EMPLOYEES and DEPARTMENT tables and select the First name, Last name and Department name of all employees.

```python
%%sql 
SELECT E.F_NAME, E.L_NAME, D.DEP_NAME
FROM EMPLOYEES E 
FULL JOIN DEPARTMENTS D ON E.DEP_ID = D.DEPT_ID_DEP;
```

##### Problem 7: Re-write the previous query but have the result set include all employee names but department id and department names only for male employees.

```python
%%sql 
SELECT E.F_NAME, E.L_NAME, D.DEPT_ID_DEP, D.DEP_NAME
FROM EMPLOYEES E 
FULL JOIN DEPARTMENTS D ON E.DEP_ID = D.DEPT_ID_DEP
AND E.SEX = 'M';
```

## Practice

##### CHICAGO DATASETS

##### Problem 1: List the case number, type of crime and community area for all crimes in community area number 18.

```python
%%sql SELECT CCD.CASE_NUMBER, CCD.PRIMARY_TYPE, CCD.COMMUNITY_AREA_NUMBER, CD.COMMUNITY_AREA_NAME
FROM CHICAGO_CRIME_DATA CCD
INNER JOIN CENSUS_DATA CD ON CCD.COMMUNITY_AREA_NUMBER = CD.COMMUNITY_AREA_NUMBER
WHERE CCD.COMMUNITY_AREA_NUMBER LIKE '%18%';
```

##### Problem 2: List all crimes that took place at a school. Include case number, crime type and community name.

```python
%%sql SELECT CCD.CASE_NUMBER, CCD.PRIMARY_TYPE, CD.COMMUNITY_AREA_NAME
FROM CHICAGO_CRIME_DATA CCD
LEFT JOIN CENSUS_DATA CD ON CCD.COMMUNITY_AREA_NUMBER = CD.COMMUNITY_AREA_NUMBER
WHERE CCD.LOCATION_DESCRIPTION LIKE '%SCHOOL%';
```

##### Problem 3: For the communities of Oakland, Armour Square, Edgewater and CHICAGO list the associated community_area_numbers and the case_numbers.

```python
%%sql SELECT CD.COMMUNITY_AREA_NUMBER, CD.COMMUNITY_AREA_NAME, CCD.CASE_NUMBER, CCD.PRIMARY_TYPE
FROM CENSUS_DATA CD
INNER JOIN CHICAGO_CRIME_DATA CCD ON CD.COMMUNITY_AREA_NUMBER = CCD.COMMUNITY_AREA_NUMBER
WHERE CD.COMMUNITY_AREA_NAME IN ('Oakland', 'Armour Square', 'Edgewater', 'CHICAGO');
```
